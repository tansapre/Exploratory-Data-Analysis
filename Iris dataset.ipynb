```python
import seaborn as sns
iris = sns.load_dataset("iris")
iris
```

```python
# Pairplots

sns.set_style("whitegrid")
sns.pairplot(iris)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.pairplot(iris,hue = 'species')
plt.show()
```
